```python
app = dash.Dash()
```

```python
external_stylesheets = ['https://codepen.io/chriddyp/pen/bWLwgP.css']
app.layout = html.Div(children = [
    
    html.H1("NYC AirBnb data", style={"textAlign": "center", "fontColor" : "blue"}),
  
    html.Div(children = [
        
        dcc.Checklist(
            id ='checklist', 
            options=[
                {'label': i, 'value': i} for i in sd.neighbourhood_group.unique()],
            value=[]
        ),
        
        
        html.Div(children = [
            dcc.Graph(
                id='graph11',
                figure={
                    'data':[
                        go.Scatter(
                        x = sd.reviews_per_month, 
                        y = sd.price, 
#                         text = sd.index,
#                         selectedpoints = 'selectedpoints',
#                         customdata = sd.index,
                        mode = 'markers',
                        marker={
                            'size': 7,
                            'opacity': 0.7,
                            'line': {'width': 0.5, 'color': 'white'}
                        }, 
                       )
                    ], 
                    'layout':go.Layout(
                        title = 'Reviews per Month Vs Price', 
                        xaxis = {'title':'Reviews Per Month'}, 
                        yaxis = {'title':'price'}, 
                        hovermode = 'closest',
                    )
                }
            )
        ], className = 'column', style={'display': 'inline-block'}),
        
        html.Div(children = [
            dcc.Graph(
                id='graph12',
                figure={
                    'data':[
                        go.Scatter(
                        x = sd.longitude, 
                        y = sd.latitude, 
                        mode = 'markers', 
                        marker={
                            'size': 7,
                            'opacity': 0.7,
                            'line': {'width': 0.5, 'color': 'white'}
                        })
                    ], 
                    'layout':go.Layout(
                        title = 'Longitude Vs Latitude', 
                        xaxis = {'title':'Longitude'}, 
                        yaxis = {'title':'Latitude'}, 
                        hovermode = 'closest'
                    )
                }
            )
        ], className = 'column', style={'align': 'right', 'display': 'inline-block'}),  
        
    ], className = 'row'), 
    
    html.Div(children = [
            
        html.Div(children = [
                dcc.Graph(
                    id='graph21',
                    figure={
                        "data": [
                            go.Pie(
                                labels = sd.neighbourhood_group.unique(), 
                                values= sd.neighbourhood_group.value_counts(),
                                hole=0.4,
                                marker={'colors': ['#EF963B', '#C93277', '#349600', '#EF533B', '#57D4F1'], 
                                    }, 
                                textinfo='label')],

                        "layout": go.Layout(title="Neighbourhood boroughs", 
#                                 margin={"l": 300, "r": 300, },
                                legend={"x": 1, "y": 0.7})
                    }

                )

            ], className = 'column', style={'display': 'inline-block'}),
            
            html.Div(children = [
                dcc.Graph(
                    id='graph22',
                    figure={
                        "data":[
                            
                            {'x':sd.neighbourhood_group.unique(), 
                             'y':sd.neighbourhood_group[sd.room_type == 'Private room'].value_counts(), 
                             'name':"Private Room", 
                             'type':'bar'}, 
                            
                            {'x':sd.neighbourhood_group.unique(), 
                             'y':sd.neighbourhood_group[sd.room_type == 'Entire home/apt'].value_counts(), 
                             'name':'Entire home/apt', 
                             'type':'bar'}, 
                            
                            {'x':sd.neighbourhood_group.unique(), 
                             'y':sd.neighbourhood_group[sd.room_type == 'Shared room'].value_counts(), 
                             'name':'Shared room', 
                             'type':'bar'}, 
                        
                        ],

                        "layout": go.Layout(title="Neighbourhood boroughs #2",)
                    }

                )

            ], className = 'column', style={'align': 'right', 'display': 'inline-block'}),

        ], className = 'row'),
    
],style={'textAlign':'center'})

@app.callback(Output('graph21', 'figure'), 
             Input('checklist', 'value'))
def update_dashboard(boroughs):
    figure={
        "data": [
            go.Pie(
                labels = sd.neighbourhood_group[sd.neighbourhood_group.unique() == boroughs], 
                values= sd.neighbourhood_group.value_counts(),
                hole= 0.4,
                marker={'colors': ['#EF963B', '#C93277', '#349600', '#EF533B', '#57D4F1'], 
                    }, 
                textinfo='label')],

        "layout": go.Layout(title="Neighbourhood boroughs", 
#                                 margin={"l": 300, "r": 300, },
                legend={"x": 1, "y": 0.7})
    }
   
    return figure
    
```

```python
if __name__ == '__main__':
    app.run_server(port = 2022)
    
```
